```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the first figure
plt.figure(1)  # Figure 1
plt.plot(x, y1, 'r-')  # Plot red sine wave
plt.title('Figure 1: Sine Wave')
plt.xlabel('X')
plt.ylabel('Sin(X)')

# Create the second figure
plt.figure(2)  # Figure 2
plt.plot(x, y2, 'b-')  # Plot blue cosine wave
plt.title('Figure 2: Cosine Wave')
plt.xlabel('X')
plt.ylabel('Cos(X)')

# Display all figures
plt.show()
```

```python

import matplotlib.pyplot as plt



def example_plot(ax):
    ax.plot([1, 2])
    ax.set_xlabel('x-label', fontsize=12)
    ax.set_ylabel('y-label', fontsize=12)
    ax.set_title('Title', fontsize=14)

fig, axs = plt.subplots(nrows=2, ncols=2, layout='constrained')

for ax in axs.flat:
    example_plot(ax)
    
```

```python

import matplotlib.pyplot as plt


fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
means = [20, 35, 30, 35]
std_devs = [2, 3, 4, 1]

fig, ax = plt.subplots()

# Bar chart
bars = ax.bar(categories, means, yerr=std_devs, capsize=5, color='lightblue', label='Mean values with CI')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Chart with Confidence Intervals')
ax.legend()

plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Example data for scatter plot
x = np.random.randint(0, 1200, size=100)  # Random data points between 0 and 1200
y = np.random.randint(0, 1200, size=100)

# Creating the scatter plot
plt.scatter(x, y)

# Adding a line from (0, 0) to (1000, 1000)
plt.plot([0, 1000], [0, 1000], 'r-', label='Line from (0,0) to (1000,1000)')  # 'r-' denotes a red line

# Setting the limits for x and y axes
plt.xlim(0, 1200)
plt.ylim(0, 1200)

# Adding labels and title
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.title('Scatter Plot with Line Through (0,0) and (1000,1000)')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```
